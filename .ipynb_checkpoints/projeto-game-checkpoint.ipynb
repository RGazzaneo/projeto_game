```python
def expandLine(line):
    return line[0]+line[5:9].join([line[1:5]*(2)]*3)+line[9:13]
```

```python
def print_board(board):
    '''Função para exibir a matriz do Sudoku
    
    Codigo modificado a partir da Fonte: https://stackoverflow.com/questions/45471152/how-to-create-a-sudoku-puzzle-in-python'''

    line0  = expandLine("╔═══╤═══╦═══╗")
    line1  = expandLine("║ . │ . ║ . ║")
    line2  = expandLine("╟───┼───╫───╢")
    line3  = expandLine("╠═══╪═══╬═══╣")
    line4  = expandLine("╚═══╧═══╩═══╝")

    symbol = " 1234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    nums   = [ [""]+[symbol[n] for n in row] for row in board ]
    
    print('                                   ← X ')
    print('  1   2   3   4   5   6   7   8   9   Y ↓')
    print(line0)
    for r in range(1,10):
        print( "".join(n+s for n,s in zip(nums[r-1],line1.split("."))),r )
        print([line2,line3,line4][(r%9==0)+(r%3==0)])
```
