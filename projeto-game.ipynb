```python
def print_start_board():
    
    '''Executa mensagem de abertura do Jogo'''
    
        input('''******************************
*                            *
*     BEM VINDO AO SUDOKU!   *
*                            *
******************************

INSTRUÇÕES:

  -> Complete todas as células usando números de 1 a 9, sem repetir os números numa mesma linha, coluna ou grade (3x3)

  -> Para preencher a celula de as coordenadas A,B e no numero a ser inserido - exemplo: (3,9,7)

  -> Para apagar a celula de as coordenadas A,B,numero a ser apagado no formato negativo - exemplo: (3,9,-7))

>>>>>>>  Para continuar aperte ENTER  <<<<<<<

''')
```

```python
def expandLine(line):
    
    '''Função auxiliar para exibir a matriz do Sudoku
    
    Codigo modificado a partir da Fonte:
    https://stackoverflow.com/questions/45471152/how-to-create-a-sudoku-puzzle-in-python'''
        
    return line[0]+line[5:9].join([line[1:5]*(2)]*3)+line[9:13]
```

```python
def print_board(board):
    
    '''Exibir a matriz do Sudoku
    
    Codigo modificado a partir da Fonte:
    https://stackoverflow.com/questions/45471152/how-to-create-a-sudoku-puzzle-in-python'''

    line0  = expandLine("╔═══╤═══╦═══╗")
    line1  = expandLine("║ . │ . ║ . ║")
    line2  = expandLine("╟───┼───╫───╢")
    line3  = expandLine("╠═══╪═══╬═══╣")
    line4  = expandLine("╚═══╧═══╩═══╝")

    symbol = " 1234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    nums   = [ [""]+[symbol[n] for n in row] for row in board ]
    
    print('                                   ← X ')
    print('  1   2   3   4   5   6   7   8   9   Y ↓')
    print(line0)
    for r in range(1,10):
        print( "".join(n+s for n,s in zip(nums[r-1],line1.split("."))),r )
        print([line2,line3,line4][(r%9==0)+(r%3==0)])
```

```python
def print_bar(i):
    
    '''Exibi barra de contagem de numeros preenchidos'''
    
    print(f'''
     *** Numeros preenchidos ***
╔═══╤═══╤═══╤═══╤═══╤═══╤═══╤═══╤═══╗
║ 1 │ 2 │ 3 │ 4 │ 5 │ 6 │ 7 │ 8 │ 9 ║
╟───┼───┼───┼───┼───┼───┼───┼───┼───╢
║ {i[0]} │ {i[1]} │ {i[2]} │ {i[3]} │ {i[4]} │ {i[5]} │ {i[6]} │ {i[7]} │ {i[8]} ║
╚═══╧═══╧═══╧═══╧═══╧═══╧═══╧═══╧═══╝''')
```

```python
def gamer_input():
    
    '''Solicita input do jogador e verifica se o input corresponde ao formato solicitado'''
    
    input_arr=input('Sua jogada:')
    input_arr=input_arr.split(',')
    
    if (len(input_arr)==3):
        if (input_arr[0].isdigit()==True)&(input_arr[1].isdigit()==True)&(input_arr[2].replace('-','').isdigit()==True):
            input_num=[int(input_arr[0]),int(input_arr[1]),int(input_arr[2])]
            return tuple(input_num)
    else:
        print('''  -> Para preencher a celula de as coordenadas X,Y e no numero a ser inserido - exemplo: (3,9,7)

  -> Para apagar a celula de as coordenadas X,Y, numero a ser apagado no formato negativo - exemplo: (3,9,-7))
  ''')
        gamer_input()
```

```python
def invalid_mov(i,board_gamer,board_empty,board_full):
    
    '''Informa ao jogador sobre jogada invalida e solicita input para continuar/resetar/parar'''

    dict={1:'>>  Essa celula não pode ser editada',
          2: '>>  Essa celula já esta ocupada, para apagar use o numero negativo',
          3:'>>  Verifique se não valor não repete na linha, coluna ou quadrado '}
    print('\n-----------------------------------------------------------------------')
    print(dict[i])
    print('-----------------------------------------------------------------------\n')
    
    board_gamer=exception(board_gamer,board_empty,board_full)
    if board_gamer==False:
        return False
    else:
        return True
```

```python
def exception(board_gamer,board_empty,board_full):
    
    '''Função auxiliar solicita input para continuar/resetar/parar'''
    
    i=input('Digite: 1=Continuar; 2=Resetar; 3=Parar')
    
    if i.isdigit()==True:
        i=int(i)
        if i==3:
            print('''
        ************************                       
        *                      *
        *      GAME OVER !     *
        *                      *
        ************************
            ''')
            print_board(board_full)
            return False
        elif i==2:
            board_gamer = board_empty
            return board_gamer
        elif i==1:
            return board_gamer
    else: 
        exception(board_gamer,board_empty,board_full)
```

```python
def varredura(board_gamer, x,y,c):
    
    '''Verifica se o numero escolhido pelo usuaria conflita com linha/coluna/quadrado'''
    
       #varredura_x(board)
    if c in board_gamer[x-1]:
        return True
        
        #varredura_y(board)
    elif c in [board_gamer[i][y-1] for i in range(9)]:
        return True
    
        #varredura_q(board)
        
    else: 
        return False
```

```python
def sudoku():
    
    board_empty = [[0, 3, 0, 9, 5, 2, 0, 7, 0],
                   [0, 9, 1, 0, 0, 0, 2, 5, 8],
                   [7, 0, 2, 0, 1, 8, 0, 3, 0],
                   [9, 0, 3, 0, 0, 6, 7, 0, 5],
                   [0, 2, 0, 7, 4, 0, 0, 6, 9],
                   [1, 0, 6, 5, 3, 0, 0, 0, 2],
                   [5, 1, 0, 4, 0, 0, 8, 2, 0],
                   [0, 0, 0, 1, 6, 7, 0, 9, 3],
                   [3, 6, 9, 8, 0, 0, 1, 0, 0]]
    
    board_full = [[4, 3, 8, 9, 5, 2, 6, 7, 1],
                  [6, 9, 1, 3, 7, 4, 2, 5, 8],
                  [7, 5, 2, 6, 1, 8, 9, 3, 4],
                  [9, 4, 3, 2, 8, 6, 7, 1, 5],
                  [8, 2, 5, 7, 4, 1, 3, 6, 9],
                  [1, 7, 6, 5, 3, 9, 4, 8, 2],
                  [5, 1, 7, 4, 9, 3, 8, 2, 6],
                  [2, 8, 4, 1, 6, 7, 5, 9, 3],
                  [3, 6, 9, 8, 2, 5, 1, 4, 7]]
    
    board_gamer = [[0, 3, 0, 9, 5, 2, 0, 7, 0],
                   [0, 9, 1, 0, 0, 0, 2, 5, 8],
                   [7, 0, 2, 0, 1, 8, 0, 3, 0],
                   [9, 0, 3, 0, 0, 6, 7, 0, 5],
                   [0, 2, 0, 7, 4, 0, 0, 6, 9],
                   [1, 0, 6, 5, 3, 0, 0, 0, 2],
                   [5, 1, 0, 4, 0, 0, 8, 2, 0],
                   [0, 0, 0, 1, 6, 7, 0, 9, 3],
                   [3, 6, 9, 8, 0, 0, 1, 0, 0]]
    empty_cell=36
    count_bar=[7,5,6,2,5,5,5,4,6]
    play=True
    
    print_start_board()
    
    while play==True:
        print_board(board_gamer)
        print_bar(count_bar)
        x,y,c=gamer_input() 
             
        if board_empty[y-1][x-1]!=0:
            play=invalid_mov(1,board_gamer,board_empty,board_full)
                    
        elif board_gamer[y-1][x-1]!=0:
            play=invalid_mov(2,board_gamer,board_empty,board_full)
                
        elif varredura(board_gamer, x,y,c)==True:
            play=invalid_mov(3,board_gamer,board_empty,board_full)
            
        else:
            if c<0:
                board_gamer[y-1][x-1]=0
                count_bar[(c*-1)-1]-=1
                empty_cell+=1
            else:
                board_gamer[y-1][x-1]=c
                count_bar[c-1]+=1
                empty_cell-=1
                if empty_cell==0:
                    play=False
                    print('''
        ************************                       
        *                      *
        *     VOCÊ GANHOU !    *
        *                      *
        ************************
            ''')
        
```
