```python
def print_start_board():
    
    '''Executa mensagem de abertura do Jogo'''
    
    input('''******************************
*                            *
*     BEM VINDO AO SUDOKU!   *
*                            *
******************************

INSTRUÇÕES:

  -> Complete todas as células usando números de 1 a 9, sem repetir os números numa mesma linha, coluna ou grade 3x3

  -> Para preencher a celula informe as coordenadas A, B e o numero a ser inserido - exemplo: (3,9,7)

  -> Para apagar a celula informe as coordenadas A, B e o numero a ser apagado no formato negativo - exemplo: (3,9,-7))

>>>>>>>  Para continuar aperte ENTER  <<<<<<<

''')
```

```python
def expandLine(line):
    
    '''Função auxiliar para exibir a matriz do Sudoku
    
    Codigo modificado a partir da Fonte:
    https://stackoverflow.com/questions/45471152/how-to-create-a-sudoku-puzzle-in-python'''
        
    return line[0]+line[5:9].join([line[1:5]*(2)]*3)+line[9:13]
```

```python
def print_board(board):
    
    '''Exibir a matriz do Sudoku
    
    Codigo modificado a partir da Fonte:
    https://stackoverflow.com/questions/45471152/how-to-create-a-sudoku-puzzle-in-python'''

    line0  = expandLine("╔═══╤═══╦═══╗")
    line1  = expandLine("║ . │ . ║ . ║")
    line2  = expandLine("╟───┼───╫───╢")
    line3  = expandLine("╠═══╪═══╬═══╣")
    line4  = expandLine("╚═══╧═══╩═══╝")

    symbol = " 1234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    nums   = [ [""]+[symbol[n] for n in row] for row in board ]
    
    print('                                   ← A ')
    print('  1   2   3   4   5   6   7   8   9   B ↓')
    print(line0)
    for r in range(1,10):
        print( "".join(n+s for n,s in zip(nums[r-1],line1.split("."))),r )
        print([line2,line3,line4][(r%9==0)+(r%3==0)])
```

```python
def invalid_mov(i,board_gamer,board_empty,board_full):
    
    '''Informa ao jogador sobre jogada invalida e solicita input para continuar/resetar/parar'''

    dict_alert={1:'>>  Essa celula não pode ser editada',
          2: '>>  Essa celula já esta ocupada, para apagar use o numero negativo',
          3:'>>  Verifique se não valor não repete na linha, coluna ou grade 3x3 '}
    print('\n-----------------------------------------------------------------------')
    print(dict_alert [i])
    print('-----------------------------------------------------------------------\n')
    
    board_gamer,play=exception(board_gamer,board_empty,board_full)
    
    if play==False:
        return (board_gamer,False)
    else:
        return (board_gamer,True)
```

```python
def exception(board_gamer,board_empty,board_full):
    
    '''Função auxiliar solicita input para continuar/resetar/parar'''
    
    i=input('Digite: 1=Continuar o jogo; 2=Resetar; 3=Parar')
    
    if i.isdigit()==True:
        i=int(i)
        if i==3:
            print('''
        ************************                       
        *                      *
        *      GAME OVER !     *
        *                      *
        ************************
            ''')
            print_board(board_full)
            return (board_full,False)
        elif i==2:
            return (board_empty,True)
        elif i==1:
            return (board_gamer,True)
    else: 
        exception(board_gamer,board_empty,board_full)
```

```python
 def game_loop(play,board_empty,board_gamer,board_full,count_bar,empty_cell):
        
        '''Executa uma rodada do jogo'''
        
        print_board(board_gamer)
        print_bar(count_bar)
        x,y,c=gamer_input() 
             
        if board_empty[y-1][x-1]!=0:
            board_gamer,play=invalid_mov(1,board_gamer,board_empty,board_full)   
            return play
        
        elif board_gamer[y-1][x-1]!=0:
            board_gamer,play=invalid_mov(2,board_gamer,board_empty,board_full)    
            return play
        
        elif varredura(board_gamer, x,y,c)==True:
            board_gamer,play=invalid_mov(3,board_gamer,board_empty,board_full)
            return play

        else:
            if c<0:
                board_gamer[y-1][x-1]=0
                count_bar[(c*-1)-1]-=1
                empty_cell+=1
            else:
                board_gamer[y-1][x-1]=c
                count_bar[c-1]+=1
                empty_cell-=1
                if empty_cell==0:
                    print('''
        ************************                       
        *                      *
        *     VOCÊ GANHOU !    *
        *                      *
        ************************
            ''')
                    return False
            
            return True
```
